# 6. Data Merging

```python
import pandas as pd
import numpy as np
```

## Concatenating data

```python
## Same Columns Name
df1 = pd.DataFrame(np.arange(16).reshape(4, 4), 
                   columns=['a', 'b', 'c','d'])


df2 = pd.DataFrame(np.arange(21, 37).reshape(4, 4), 
                   columns=['a', 'b', 'c','d'])
df1
```

```python
df2
```

```python
pd.concat([df1, df2]) ## Concat 1
```

```python
pd.concat([df1, df2], axis = 1) ## Concat 2
```

```python
## Different Columns Name
df1 = pd.DataFrame(np.arange(9).reshape(3, 3), 
                   columns=['a', 'b', 'c'])


df2 = pd.DataFrame(np.arange(30, 39).reshape(3, 3), 
                   columns=['d', 'b', 'c'])
df1
```

```python
df2
```

```python
# will be Nan
pd.concat([df1, df2])
```

```python
# or You can use append
df1.append(df2)
```

```python
# or You can use append (and ignore the index)
df1.append(df2, ignore_index=True)
```

```python
pd.concat([df1, df2], axis=1)
```

```python
## different Index
df3 = pd.DataFrame(np.arange(30, 36).reshape(3, 2), 
                   columns=['a', 'd'], 
                   index=[2, 3, 4])
df3
```

```python
df1
```

```python
# concat them. Alignment is along row labels
# columns first from df1 and then df3, with duplicates.
# NaN filled in where those columns do not exist in the source
pd.concat([df1, df3], axis=1)
```

## Left/Right Join

like vlookup in excel

```python
PATH_DATA = '../mydata/employee_data/'
```

```python
employee_votes = pd.read_csv(PATH_DATA+'votes.csv')
employee_churn = pd.read_csv(PATH_DATA+'churn.csv')
```

```python
col = 'voteDate'
len(employee_votes[col].unique())==len(employee_votes)
```

```python
employee_votes.head()
```

```python
employee_churn.head()
```

```python
'A'+'-'+'becak'
```

```python
## unique id - create primary key
employee_votes['emp-company']=employee_votes['employee'].astype(str) + '-' + employee_votes['companyAlias']
employee_churn['emp-company']=employee_churn['employee'].astype(str) + '-' + employee_churn['companyAlias']
```

```python
employee_votes.head()
```

```python
employee_churn.head()
```

```python
df0 = employee_churn[['emp-company','stillExists']]
df0.head()
```

```python
join_churn_to_votes_left = pd.merge(employee_votes,df0,
                                           on = 'emp-company', how = 'left') ## employee_votes <-- df0
join_churn_to_votes_right = pd.merge(df0,employee_votes,
                                           on = 'emp-company', how = 'right') ## df0 --> employee_votes
```

```python
join_churn_to_votes_left.head()
```

```python
join_churn_to_votes_right.head()
```

```python
join_churn_to_votes_left.shape,join_churn_to_votes_right.shape
```

```python
## use multi columns
df1 = employee_churn[['employee','companyAlias','stillExists']]

join_churn_to_votes_left2 = pd.merge(employee_votes,df1,
                                           on = ['employee','companyAlias'], how = 'left') ## employee_votes <-- df1
join_churn_to_votes_right2 = pd.merge(df1,employee_votes,
                                           on = ['employee','companyAlias'], how = 'right') ## df1 --> employee_votes
```

```python
join_churn_to_votes_left2.head()
```
