Correlation of daily order number vs Avg Delivery Fee

```python
df_food = pd.read_csv(PATH_DATA+'food_delivery_datasets.csv')
```

```python
df_food.head()
```

```python
## Datetime issue
df_food['date_time'] = pd.to_datetime(df_food['date_time'])
df_food['date_c'] = df_food['date_time'].dt.strftime('%Y-%m-%d')
```

```python
df_food.head()
```

```python
group_typ = df_food.groupby(['date_c'])
group_typ_measure1 = group_typ.agg({'order_id':np.size,'delivery_fee':np.mean})
group_typ_measure1.corr()
```

```python
group_typ
```

```python
group_typ_measure1
```

compare the Avg Food Price of merchant with high eta vs merchant with low eta

```python
df_food['eta_seconds'].quantile(0.8)
```

```python
2300/60
```

```python
df_food['eta_types'] = np.where(df_food['eta_seconds']<2300, 'Low', 'High')
```

```python
df_food.head()
```

```python
df_compare_food_price = df_food.groupby(['date_c','eta_types'])['food_price'].mean().unstack()
```

```python
df_compare_food_price
```

```python
df_compare_food_price['is_higher'] = (df_compare_food_price['High']>df_compare_food_price['Low']).astype(int)
```

```python
df_compare_food_price
```

impute the rating with the previous day rating of the merchant, what is the overall average?

```python
df_food.sample(5)
```

```python
df_resto_rating = df_food.groupby(['date_c','resto_id'])['rating'].mean().unstack().fillna(4).stack().reset_index()
df_resto_rating.rename(columns = {0:'rating_imp','date_c': 'yesterday'}, inplace=True)
```

```python
df_resto_rating.head()
```

```python
df_resto_rating['date_c'] = pd.to_datetime(df_resto_rating['yesterday']) + pd.to_timedelta(1, 'D')
```

```python
df_resto_rating
```

```python
df_food['date_c'] = pd.to_datetime(df_food['date_c'])
df_food2 = pd.merge(df_food, df_resto_rating, on = ['date_c','resto_id'], how = 'left')
```

```python
df_food2['rating_new'] = df_food2['rating'].fillna(df_food2['rating_imp'])
```

```python
df_food2.head()
```

```python
df_food2['rating_new'].mean()
```
